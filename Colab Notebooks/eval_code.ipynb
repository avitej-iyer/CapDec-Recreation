```python
import json
import os
from collections import defaultdict
from google.colab import files

references_output_dir = "/content/bleu-rouge-meteor-cider-spice-eval4imagecaption/example"
captions_output_dir = "/content/bleu-rouge-meteor-cider-spice-eval4imagecaption/example"

os.makedirs(references_output_dir, exist_ok=True)
os.makedirs(captions_output_dir, exist_ok=True)

# Upload references.json
print("Upload references.json:")
references_uploaded = files.upload()
references_file_name = list(references_uploaded.keys())[0]

# Upload captions.json
print("Upload captions.json:")
captions_uploaded = files.upload()
captions_file_name = list(captions_uploaded.keys())[0]

# Load references.json
with open(references_file_name, 'r') as f:
    raw_references = json.load(f)

# Load captions.json
with open(captions_file_name, 'r') as f:
    raw_captions = json.load(f)

# Extract valid image IDs from captions.json
caption_image_ids = {str(entry["image_id"]) for entry in raw_captions}

# Filter references.json to include only image IDs present in captions.json
references = {"images": [], "annotations": []}
seen_image_ids = set()
annotation_id = 0  # Unique ID for annotations

for entry in raw_references:
    image_id = str(entry["image_id"])
    if image_id in caption_image_ids:  # Keep only matching image IDs
        if image_id not in seen_image_ids:
            references["images"].append({"id": image_id})
            seen_image_ids.add(image_id)
        references["annotations"].append({
            "image_id": image_id,
            "id": annotation_id,
            "caption": entry["caption"]
        })
        annotation_id += 1

# Build captions.json structure with exactly one caption per image_id
captions = {}
for entry in raw_captions:
    image_id = str(entry["image_id"])
    if image_id not in captions:
        captions[image_id] = entry["caption"]  # Take the first caption for each image_id

# Format captions.json as a list of dictionaries
captions_list = [{"image_id": img_id, "caption": cap} for img_id, cap in captions.items()]

# Save processed references.json
references_output_file = os.path.join(references_output_dir, "references.json")
with open(references_output_file, 'w') as f:
    json.dump(references, f, indent=2)

# Save processed captions.json
captions_output_file = os.path.join(captions_output_dir, "captions.json")
with open(captions_output_file, 'w') as f:
    json.dump(captions_list, f, indent=2)

print(f"Processed references.json saved to: {references_output_file}")
print(f"Processed captions.json saved to: {captions_output_file}")
```
